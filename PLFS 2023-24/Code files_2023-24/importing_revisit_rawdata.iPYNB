```python
# Loading HH and Person revisit files
df_hhrv = pd.read_fwf(filepath_or_buffer=file_path_hhrv, colspecs=col_specs_hhrv, names = col_names_hhrv, dtype = str)
df_perrv = pd.read_fwf(filepath_or_buffer=file_path_perrv, colspecs=col_specs_perrv, names = col_names_perrv, dtype = str)
```

```python
# Creating unique IDs for merging

#HHRV - Household ID
df_hhrv["hhid"] = df_hhrv['hhvar3'] + df_hhrv['hhvar4'] + df_hhrv['hhvar5'] + df_hhrv['hhvar13']+ df_hhrv['hhvar14']+ df_hhrv['hhvar15']+ df_hhrv['hhvar16']

##PERRV - Household ID
df_perrv["hhid"] = df_perrv['pvar3'] + df_perrv['pvar4'] + df_perrv['pvar5'] + df_perrv['pvar13'] + df_perrv['pvar14'] + df_perrv['pvar15'] + df_perrv['pvar16']

##PERRV - Person ID
df_perrv["perid"] = df_perrv['pvar3'] + df_perrv['pvar4'] + df_perrv['pvar5'] + df_perrv['pvar13'] + df_perrv['pvar14'] + df_perrv['pvar15'] + df_perrv['pvar16'] + df_perrv['pvar17']
```

```python
#Merging HHV1 and PERV1
df_merged = pd.merge(df_hhrv,df_perrv, on = "hhid", how="outer",indicator= True)
df_merged["_merge"].value_counts() #Checking for unmerged rows
```
