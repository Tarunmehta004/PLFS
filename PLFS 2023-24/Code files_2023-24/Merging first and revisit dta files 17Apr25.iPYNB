# Import libraries

```python
import pandas as pd
import numpy as np
```

```python
# Setting options to display all columns in a df
pd.set_option("display.max_columns", None)
```

# Import dataframe

```python
# Path for merged HH and Per first visit dta file
path_fv = r"..\Extracted_files\HHV1_PERV1_merged.dta"

# Path for merged HH and Per revisit dta file
path_rv = r"..\Extracted_files\HHRV_PERRV_merged.dta"
```

```python
# Loading dataframes
df_fv = pd.read_stata(path_fv).copy()
df_rv = pd.read_stata(path_rv).copy()
```

```python
# Changing vartypes to numeric
df_fv[['pvar134','pvar138']] = df_fv[['pvar134','pvar138']].apply(pd.to_numeric)
df_rv[['pvar134','pvar138']] = df_rv[['pvar134','pvar138']].apply(pd.to_numeric)
```

```python
# Separating urban and rural first visit samples
df_fv_urban = df_fv.loc[df_fv['pvar5'] == 'urban',:]
df_fv_rural = df_fv.loc[df_fv['pvar5'] == 'rural',:]
```

```python
df_rv.head()
```

```python
# Merging first and revisits for urban sample. Note: Revisits only happen for urban areas
df_merged_urban = pd.concat([df_fv_urban,df_rv], join = 'inner', ignore_index=True).copy()

# Merging all urban visits and rural first visits
df_merged = pd.concat([df_merged_urban, df_fv_rural], join = 'inner', ignore_index=True).copy()
```

```python
# Annual weights for df_merged_urban (Note: Weights are for a quarter, hence division by 4. Each quarter has 4 urban panels, hence further division by 4)
df_merged_urban['weights'] = np.where(df_merged_urban['pvar136'] == df_merged_urban['pvar137'], df_merged_urban['pvar138']/(100*4*4),df_merged_urban['pvar138']/(200*4*4))
```

```python
# Annual weights for df merged

## Assigning weights to rural and urban areas together
df_merged['weights'] = np.where(df_merged['pvar136'] == df_merged['pvar137'], df_merged['pvar138']/(100*4),df_merged['pvar138']/(200*4))

## Adjusting for urban panels in each quarter
df_merged['weights'] = df_merged[['weights', 'pvar5']].apply(lambda row: row['weights']/4 if row['pvar5'] == '2' else row['weights'], axis = 1) 
```

## Urban sample (First visit + Three revisits)

```python
# Creating a duplicate df
trial = df_fv.copy()

# Filtering for regular salaried workers based on CWS
trial = trial[trial['pvar131'].isin(["31",'71','72'])]

# Weighted avg salary of people in regular salaried job
print((trial['pvar134']*trial['weights']).sum()/trial['weights'].sum())

# Weighted avg salary of people in regular salaried job - grouped by sector
print(trial.groupby('pvar5').apply(lambda table: (table['pvar134']*table['weights']).sum()/table['weights'].sum(), include_groups =False))
```
