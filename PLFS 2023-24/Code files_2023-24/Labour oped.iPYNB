## Changing dataformats

```python
# Converting age, status code (principal), status code (subsidiary), and status code (usual) to numeric
df[["pvar20","pvar32", "pvar43", "usual_status_code", 'pvar134']] = df[["pvar20","pvar32", "pvar43", "usual_status_code", 'pvar134']].apply(pd.to_numeric)
```

# Analysis

```python
# Employment in coal power plants (NIC = 35102)

# Assigning NIC code based on usual status
df["usual_status_ind"] = np.where((df['pvar32'] >=81) & (df['pvar43'].notna()),   # Condition for subsidiary code to be not NA and principal to be more than or equal to 81
                                    df['pvar44'],
                                    df['pvar33'])

# Total number of workers in coal plants
df.loc[df['usual_status_ind'] == '35102', 'weights'].sum()
```

# Creating new columns

## Usual status industry

```python
#Creating a new column with 2 digit NIC codes based on usual status

#Assigning NIC code based on usual status
df["usual_status_ind"] = np.where((df['pvar32'] >=81) & (df['pvar43'].notna()),   # Condition for subsidiary code to be not NA and principal to be more than or equal to 81
                                    df['pvar44'],
                                    df['pvar33'])

#Keeping only the first two NIC digits
df["usual_status_ind"] = df["usual_status_ind"].astype(str).str[:2]
df["usual_status_ind"] = df["usual_status_ind"].apply(pd.to_numeric)
```

## Usual status NCO

```python
#Creating a new column with 1 digit NCO codes based on usual status

#Assigning NCO code based on usual status
df["usual_status_nco"] = np.where((df['pvar32'] >=81) & (df['pvar43'].notna()),   # Condition for subsidiary code to be not NA and principal to be more than or equal to 81
                                    df['pvar45'],
                                    df['pvar34'])

#Keeping only the first digit
df["usual_status_nco"] = df["usual_status_nco"].astype(str).str[:1]
df["usual_status_nco"] = df["usual_status_nco"].apply(pd.to_numeric)
```

## Ususal status contract

```python
#Creating a new column with codes for contract type

#Assigning contract type based on usual status
df["usual_status_contract"] = np.where((df['pvar32'] >=81) & (df['pvar43'].notna()),   # Condition for subsidiary code to be not NA and principal to be more than or equal to 81
                                    df['pvar49'],
                                    df['pvar39'])
```

## Usual status social security

```python
#Creating a new column with codes for social security

#Assigning social security based on usual status
df["usual_status_social_security"] = np.where((df['pvar32'] >=81) & (df['pvar43'].notna()),   # Condition for subsidiary code to be not NA and principal to be more than or equal to 81
                                    df['pvar51'],
                                    df['pvar41'])
```

## Usual status location of workplace

```python
df["usual_status_workplace"] = np.where((df['pvar32'] >=81) & (df['pvar43'].notna()),   # Condition for subsidiary code to be not NA and principal to be more than or equal to 81
                                    df['pvar46'],
                                    df['pvar36'])
```

## Usual status informal

```python
df['usual_status_informal'] = np.where((df['usual_status_contract'] == '1') | (df['usual_status_social_security'] == '8') | (df['usual_status_code'].isin([11,12,21])), 1, 0)
```

# Analysis

## Share informal employment

```python
trial = df.groupby('usual_status_informal', dropna=False).apply(lambda table: pd.Series({
    'sample_size': table['weights'].shape[0],
    'population_size': table['weights'].sum()
}), include_groups = False)
trial
```

## Share blue collar workers

```python
informal_mask = df['usual_status_informal'] == 1
blue_collar_mask = df['usual_status_nco'].isin(range(5,9+1))
non_agri_blue_collar_mask = df['usual_status_nco'].isin([5,7,8,9])
```

```python
# Total informal workers
pd.Series({'sample size': df.loc[informal_mask,:].shape[0],
           'population size': df.loc[informal_mask,'weights'].sum()}).to_frame()
```

```python
# Total blue collar workers
pd.Series({'sample size': df.loc[blue_collar_mask,:].shape[0],
           'population size': df.loc[blue_collar_mask,'weights'].sum()}).to_frame()
```

```python
# Total informal, blue collar workers
pd.Series({'sample size': df.loc[blue_collar_mask & informal_mask,:].shape[0],
           'population size': df.loc[blue_collar_mask & informal_mask,'weights'].sum()}).to_frame()
```

```python
# Total non-agri,informal, blue collar workers
pd.Series({'sample size': df.loc[non_agri_blue_collar_mask & informal_mask,:].shape[0],
           'population size': df.loc[non_agri_blue_collar_mask & informal_mask,'weights'].sum()}).to_frame()
```
